## 3.1 MNIST
MNIST 데이터셋은 총 70,000개의 손 글씨 숫자 이미지를 모은 데이터입니다. 
각 이미지 데이터는 0부터 9까지의 숫자 중 하나가 적혀 있으며, 각 이미지는 28*28=784개의 픽셀로 구성됩니다.
이 데이터셋은 머신러닝 분류 문제의 가장 대표적인 예제 입니다. 

```python
## 데이터 내려받기
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version = 1, as_frame=False)  # sklearn에서 불러오는 데이터는 딕셔너리 형태로 저장된다.
mnist.keys()
```

__각 key의 의미__ 

- `data` : 784개의 픽셀 데이터값
- `target` : 데이터의 라벨 (0 - 9 중 하나의 숫자)
- `frame` : `data` 뒤에 `target` 값이 붙어있는 형태. 즉, 784개의 픽셀값과 맨 뒤의 label값 총 785개의 값을 가짐.
- `DESCR` : 데이터셋에 대한 설명(DESCRibe)
- `feature_names` : 데이터의 각 특성의 이름을 담고있음. MNIST의 경우 pixel1~pixel784

__SGDClassifier ?__

Machine Learning의 Pipeline은 
1. 데이터 전처리
2. 데이터를 모델(ex. SVM, Logistic Regression, Perceptron 등)에 입력해 예측함.
3. 예측의 정확도를 Loss function으로 평가
4. Loss function이 최소화 되도록 모델의 parameter를 조정 (주로 Gradient Descent의 방법 등을 사용함)
으로 진행됩니다.

`SGDClassifier`는 SVM, Logistic Regression, Perceptron 등 classification에 사용되는 linear model들의 loss를 Stochastic Gradient Descent 기법으로 최적화시키는 분류기를 말합니다.

즉 위 Pipeline에서 2번을 Linear 모델 군으로 선택하고 그에 알맞는 loss function을 Stochastic Gradient Descent 방법으로 최적화시키는 과정을 한 번에 수행해주는 것이 `SGDClassifier`입니다.


`SGDClassifier`에 loss 변수를 지정해줄 수 있는데, 이 loss를 어떤 것으로 지정하느냐에 따라 SVM을 할지, Logistic Regression을 할지, Perceptron을 할지 결정된다고 합니다.

`SGDClassifier`의 default로 설정된 loss 는 `loss = 'hinge'` 이며, 이때는 SVM 모델을 사용하게 된다고 합니다.
그 외에 `loss = 'log'` 로 설정하면 Logistic regression을, `loss = 'perceptron'`으로 설정하면 Perceptron 모델을 수행한다고 합니다. 

`SGD`에 대해서는 ... 다음 시간이 될 때에...



```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state = 42)
sgd_clf.fit(X_train, y_train_5)
```

```python
## 계층적 샘플링을 한 뒤 3-fold 교차검증

from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
# 클래스별 비율이 유지되도록 폴드를 만들기 위해 계층적 샘플링을 수행함.

```

```python
## 3-fold 교차 검증

from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_train, y_train_5, cv = 3, scoring = 'accuracy')
```

```python
## '5 아님' 클래스 분류기
from sklearn.base import BaseEstimator

## BaseEstimator 모듈에 classifier를 만들기 위한 재료가 들어있다.
## 그래서 나만의 Classifier()의 괄호 안에 BaseEstimator를 집어 넣으면
## BaseEstimator에 준비된 재료들을 사용할 수 있다.
## 가령, fit, transform 등을 작동시키기 위한 준비들이 BaseEstimator에 잘 되어 있어서 설정만 해주면 된다.

class Never5Classifier(BaseEstimator):
    def fit(self, X, y=None):
        return self             # BaseEstimator에 fit에 대한 준비가 잘 되어 있어서 세팅만 해주면 된다.
    def predict(self, X):
        return np.zeros((len(X),1), dtype = bool)   # 이 코드는 내가 뭘 넣든 모두 5가 아니라고 예측하겠다는 뜻이다.
```

```python
never_5_clf = Never5Classifier()
cross_val_score(never_5_clf, X_train, y_train_5, cv = 3, scoring = 'accuracy')

# 무지성으로 5가 아니라고 했더니 90퍼센트의 정확도를 보인다. 
# 왜냐면,,, 애초에 데이터 중에 5인 것이 10퍼센트밖에 없으니까!
# 그렇지만 이렇게 되면 불균형한 데이터셋을 다룰 때 문제가 생긴다!
```

### 3.3.3 정밀도와 재현율

오차 행렬이 많은 정보를 제공해주기도 하지만, 가끔 더 요약된 지표가 필요할 때도 있다.
그 중 하나가 양성 예측의 정확도(예측을 참이라고 했을 때, 정답도 참일 확률)를 나타내는 __정밀도(Precision)__ 이다.

$$ Precision = \frac{TP}{TP+FP}$$

또 다른 지표는 분류기가 정확하게 감지한 양성 샘플의 비율 (정답이 참일때, 예측도 참이라고 할 확률)을 나타내는 __재현율(Recall)__ 과 함께 사용하는 것이 일반적이다. 

$$ Recall = \frac{TP}{TP+FN}$$

```python
## Precision/Recall 구현

from sklearn.metrics import precision_score, recall_score

print(precision_score(y_train_5, y_train_pred),
    recall_score(y_train_5, y_train_pred))
```

정밀도와 재현율을 **$F_1$ score** 라는 하나의 숫자로 만들면 편리할 때가 많다. 특히 두 분류기를 비교할 때 그렇다. **$F_1$ score**는 정밀도와 재현율의 __조화 평균__ 이다.

$$ F_1 = \frac{2}{\frac{1}{Precision} + \frac{1}{Recall}} $$

```python
from sklearn.metrics import f1_score
f1_score(y_train_5 , y_train_pred)
```

### 3.3.5 ROC 곡선

__ROC(receiver operating characteristic)__ 곡선도 이진 분류에 널리 사용되는 지표다. 
ROC 커브는 __거짓 양성 비율 (FPR)__ 에 대한 __진짜 양성 비율(TPR)__ 의 곡선이다.
FPR은 정답이 거짓인데 참이라고 예측한 샘플의 비율이다.
TPR은 정답이 참인데 참이라고 예측한 샘플의 비율 (재현율)이다.
FPR은 1에서 정답이 거짓인데 거짓이라고 예측한 샘플의 비율인 __진짜 음성 비율(TNR)__ 을 뺸 값이다. 
TNR을 특이도라고도 한다. 따라서, ROC커브는 재현율에 대한 1-특이도 그래프이다.

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)

def plot_roc_curve(fpr, tpr, label = None):
    plt.plot(fpr, tpr, linewidth = 2, label = label)
    plt.plot([0,1],[0,1], 'k--')
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.grid()

plot_roc_curve(fpr, tpr)
plt.show()
```

여기에서도 트레이드오프가 있다.
TPR이 높을수록 FPR이 늘어난다. 좋은 분류기는 최대한 대각선에서 멀어져 있어야 한다.
즉, 곡선 아래 면적(AUC)의 값이 클 수록 좋다.
완벽한 분류기는 AUC가 1에 가깝고, 완전 랜덤 분류기는 0.5에 가깝다. 


```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_5, y_scores)
```

ROC 커브와 PR 커브 중 어떤 것을 사용해야 할지 궁금할 수 있다.
일반적으로는 양성 클래스가 드물거나, 거짓 음성보다 거짓 양성이 더 중요할 때 PR 커브를 사용하고 그렇지 않으면 ROC 커브를 사용한다.

8에 밝은 부분이 많은걸 보니 8은 상당히 잘못 분류된것 같다.

예를 들어, 8처럼 보이지만 실제 8은 아닌 숫자의 훈련 데이터를 더 많이 모아서 실제 8과 구분하도록 분류기를 학습시킬 수 있다.
또는 분류기에 도움이 될 만한 특성을 더 찾아볼 수 있다. 예를 들어 동심원의 수를 세는 알고리즘 같은 것이다. ( *TDA가 정확히 이런 일을 해줄 수 있다*)



```python
import matplotlib

# 숫자 그림을 위한 추가 함수
def plot_digits(instances, images_per_row=10, **options):
    size = 28       # 이미지의 사이즈 (가로=세로=28)
    images_per_row = min(len(instances), images_per_row)
    # 입력한 이미지와 각 row당 이미지수 중 작은 값을 반환한다.
    images = [instance.reshape(size,size) for instance in instances]
    # 이미지를 28*28 로 reshape해서 저장해둠.
    n_rows = (len(instances) - 1) // images_per_row + 1
    # 열의 수
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    # 공백칸
    images.append(np.zeros((size, size * n_empty)))
    # 공백칸을 추가함
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = matplotlib.cm.binary, **options)
    plt.axis("off")
```

```python
cl_a, cl_b = 3, 5
X_aa = X_train[(y_train == cl_a) & (y_train_pred == cl_a)]
X_ab = X_train[(y_train == cl_a) & (y_train_pred == cl_b)]
X_ba = X_train[(y_train == cl_b) & (y_train_pred == cl_a)]
X_bb = X_train[(y_train == cl_b) & (y_train_pred == cl_b)]

plt.figure(figsize = (8,8))
plt.subplot(221); plot_digits(X_aa[:25], images_per_row=5)
plt.subplot(222); plot_digits(X_ab[:25], images_per_row=5)
plt.subplot(223); plot_digits(X_ba[:25], images_per_row=5)
plt.subplot(224); plot_digits(X_bb[:25], images_per_row=5)
plt.show()
```

대부분의 잘못 분류된 이미지는 확실히 에러인 것 같고 분류기가 실수한 이유를 이해하기 어렵다. 
원인은 선형 모델인 `SGDClassifier`를 사용했기 때문이다. 
선형 모델은 클래스마다 픽셀에 가중치를 할당하고 새로운 이미지에 대해 단순히 픽셀 강도의 가중치 합을 클래스 점수로 계산한다. 따라서 3과 5는 몇 개의 픽셀만 다르기 때문에 모델이 쉽게 혼동하게 된다.
